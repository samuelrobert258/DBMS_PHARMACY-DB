1) WAQ  TO DISPLAY CUSTOMER DETAILS WHO ARE FROM BANDRA OR VASAI

2) WAQ TO DISPLAY THE DETAILS OF DRUGS WHOSE PRICE IS GREATER THAN 100.

```python
q2='''
SELECT *
FROM INVENTORY
WHERE PRICE > 100
'''
table = pd.read_sql_query(q2, conn)
table
```

3) WAQ TO DISPLAY THE NUMBER OF ITEMS WHICH HAS BEEN DELIVERED 

```python
q3='''
SELECT ITEM_STATUS, COUNT(ITEM_STATUS)
FROM ITEM 
GROUP BY ITEM_STATUS
'''
table = pd.read_sql_query(q3, conn)
table
```

4) WAQ TO DISPLAY DETAILS WITH UNIT PRICE IN DESCENDING ORDER

```python
q4='''
SELECT *
FROM INVENTORY
ORDER BY PRICE DESC
'''
table = pd.read_sql_query(q4, conn)
table
```

5) WAQ TO DISPLAY DISCOUNTED PRICE ALONG WITH ALL ATTRIBUTES OF ORDERS.

```python
q5='''
SELECT O_ID, O_DATE, O_QUANTITY, TOTAL_PRICE, TOTAL_PRICE - (TOTAL_PRICE * 0.1) AS DISCOUNTED_PRICE
FROM ORDERS

'''
table = pd.read_sql_query(q5, conn)
table
```

6) WAQ TO FIND THE DETAILS DRUG WHICH HAS NOT BEEN DELIVERED.

```python
q6='''
SELECT  U.ITEM_ID, U.EXPIRY, U.ITEM_STATUS, I.DRUG_ID, I.DRUG_NAME
FROM INVENTORY AS I NATURAL JOIN ITEM AS U
WHERE U.ITEM_STATUS = 'NOT DELIVERED'
'''
table = pd.read_sql_query(q6, conn)
table
```

7) WAQ TO DISPLAY THE DETAILS OF THE SUPPLIER WHO SUPPLIES THE ITEMS WHICH HAVE LESS THAN 20 UNITS IN THE INVENTORY.

```python
q7='''
SELECT *
FROM SUPPLIER
WHERE S_ID IN
         (SELECT  S_ID
           FROM  ITEM
           WHERE DRUG_ID IN 
                      (SELECT  DRUG_ID
                        FROM  INVENTORY
                        WHERE  UNIT  <  20))
'''
table = pd.read_sql_query(q7, conn)
table
```

8) WAQ TO DISPLAY ITEM DETAILS SUPPLIED BY THE SUPPLIER HAVING SUPPLIER ID 8796.

```python
q8='''
SELECT  *
FROM ITEM AS I NATURAL INNER JOIN SUPPLIER AS S
GROUP BY S_NAME 
HAVING  S_ID = 6789
'''
table = pd.read_sql_query(q8, conn)
table
```

9) WAQ to Simulate Full Outer Join between ORDERS and EMPLOYEE.

```python
q9='''
SELECT  *
FROM ORDERS AS O NATURAL LEFT OUTER JOIN EMPLOYEE AS E

UNION ALL

SELECT  *
FROM   EMPLOYEE AS E NATURAL LEFT OUTER JOIN ORDERS AS O
'''
table = pd.read_sql_query(q9, conn)
table
```

10) WAQ TO DISPLAY THE CUSTOMER WHO HAS ORDERED PARACETAMOL FROM THE EMPLOYEE. 

```python
q10='''
SELECT  C.C_ID, C.C_NAME, C.C_ADDRESS, C.C_CONTACT_NO, V.DRUG_NAME, E.EMP_NAME
FROM  CUSTOMER AS C ,  ORDERS AS O,  ITEM AS I, INVENTORY AS V, EMPLOYEE AS E
WHERE I.O_ID = O.O_ID
AND      O.C_ID = C.C_ID
AND      I.DRUG_ID = V.DRUG_ID
AND      E.EMP_ID = O.EMP_ID
AND      V.DRUG_NAME = 'PARACETAMOL'
'''
table = pd.read_sql_query(q10, conn)
table
```
